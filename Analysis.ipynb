```python
# Make Word Cloud for Data Science
df1 = df[df['job_cat'] == 'data scientist']
text = " ".join(review for review in df1.description)
tokenizer = RegexpTokenizer(r'\w+')
t1 = tokenizer.tokenize(text)
t1 = [x.lower() for x in t1]

words = []
for counter, x in enumerate(t1):
    if counter % 10000 == 0:
        print(f'{counter} out of {len(t1)}')
    if x not in stopwords.words():
        words.append(x)
```

```python
from nltk.probability import FreqDist
fdist = FreqDist(words)
```

```python
fdist.most_common(25)
```

```python
text = " ".join(review for review in words)
from wordcloud import WordCloud
wordcloud = WordCloud(max_font_size=50, max_words=200, background_color="black").generate(text)
fig = plt.figure(figsize = (12, 8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
fig.savefig('data_science.jpg')
```

```python
# make word cloud for data analyst 
# Make Word Cloud for Data Science
df_da = df[df['job_cat'] == 'data analyst']
text_da = " ".join(review for review in df_da.description)
tokenizer = RegexpTokenizer(r'\w+')
t_da = tokenizer.tokenize(text_da)
t_da = [x.lower() for x in t_da]

words_da = []
for counter, x in enumerate(t_da):
    if counter % 10000 == 0:
        print(f'{counter} out of {len(t_da)}')
    if x not in stopwords.words():
        words_da.append(x)
```

```python
fdist_da = FreqDist(words_da)
```

```python
fdist_da.most_common(25)
```

```python
text_da = " ".join(review for review in words_da)
wordcloud = WordCloud(max_font_size=50, max_words=200, background_color="black").generate(text_da)
fig = plt.figure(figsize = (12, 8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
fig.savefig('data_analyst.jpg')
```

```python
da_25 = fdist_da.most_common(25)
ds_25 = fdist.most_common(25)
```

```python
fig = plt.figure(figsize = (12, 8))


ax1 = plt.subplot(121)
words = [word[0] for word in da_25]
y_pos = np.arange(len(words))
ax1.barh(y_pos, [word[1] for word in da_25])
plt.yticks(y_pos)
ax1.set_yticklabels(words)
plt.gca().invert_yaxis()
plt.title('25 Most Common Data Analyst Words')

ax2 = plt.subplot(122)
words = [word[0] for word in ds_25]
y_pos = np.arange(len(words))
ax2.barh(y_pos, [word[1] for word in ds_25])
plt.yticks(y_pos)
ax2.set_yticklabels(words)
plt.gca().invert_yaxis()
plt.title('25 Most Common Data Science Words');

plt.savefig('scientist_vs_analyst.jpg')
```
